```python
# -*- coding: utf-8 -*-
"""
Created on Sun Jan 28 16:15:30 2024

@author: kaabir
"""
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(50)

# Constants
num_chromatin_classes = 1
num_markers = 5
num_cells = 7  # Number of cells for each marker and chromatin class

# Generate random intensity data for each marker and chromatin class
intensity_data = np.random.rand(num_markers, num_chromatin_classes, num_cells)

# Simulate enrichment or depletion by adding or subtracting random values
fold_change_data = intensity_data + np.random.normal(loc=0.1, scale=0.2, size=(num_markers, num_chromatin_classes, num_cells))

# Calculate log2 fold change relative to a random distribution
log2_fold_change_data = np.log2(fold_change_data / intensity_data)

# Create a separate heatmap for each marker
for marker_idx in range(num_markers):
    plt.figure(figsize=(4, 3))
    sns.heatmap(log2_fold_change_data[marker_idx], annot=True, cmap='coolwarm', fmt='.2f', cbar_kws={'label': 'Log2 Fold Change'})
    plt.xlabel('Chromatin Class')
    plt.ylabel(f'Marker {marker_idx + 1}')
    plt.title(f'Heatmap of Log2 Fold Change for Marker {marker_idx + 1} Relative to Random Distribution')
    plt.show()
```
