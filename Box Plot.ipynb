```python
# Another way for grouped box plot
import numpy as np
import matplotlib.pyplot as plt

# Random test data
np.random.seed(19680801)
all_data = [np.random.normal(0, std, size=100) for std in range(1, 4)]
all_data2 = [np.random.normal(0, std, size=100) for std in range(1, 4)]
fig, ax = plt.subplots(figsize=(8, 6))

# Boxplot
positions = [1, 3, 5]  # Adjust positions for gap
colors = ['mediumseagreen', 'darkcyan', 'indianred']

for i, (data, data2) in enumerate(zip(all_data, all_data2)):
    box = ax.boxplot([data, data2], positions=[positions[i], positions[i]+0.6], widths=0.4, patch_artist=True)
    for patch, color in zip(box['boxes'], colors):
        patch.set_facecolor(color)

# Customizing the plot
ax.set_title('Fixed E-Nuc. H3K.- Markers ') # (Non-Shrinking)
categories = ['Ctrl', 'Actin']  # Category labels
ax.set_xticks([2, 4])  # Adjust x-axis ticks to the midpoint of each group
ax.set_xticklabels(categories)
ax.set_ylabel(r'$Intensity \ (A.U)$')
ax.set_xlabel('(Ctrl vs Actin)')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.grid(which='major', color='#949494', linestyle='--')
plt.grid(which='minor', color='#d6d6d6', linestyle=':')
#plt.savefig('Frequency Distribution of Minimum Distance to the Boundary_.png', dpi=400, bbox_inches="tight", pad_inches=0.1)
plt.show()
```

```python
# Matplotlib guide for simple boxplot
import matplotlib.pyplot as plt
import numpy as np

# Random test data
np.random.seed(19680801)
all_data = [np.random.normal(0, std, size=100) for std in range(1, 4)]
labels = ['x1', 'x2', 'x3']

fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))

# rectangular box plot
bplot1 = ax1.boxplot(all_data,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax1.set_title('Rectangular box plot')

# fill with colors
colors = ['mediumseagreen', 'darkcyan', 'indianred']

for patch, color in zip(bplot1['boxes'], colors):
        patch.set_facecolor(color)

# adding horizontal grid lines

ax1.yaxis.grid(True)
ax1.set_xlabel('Three separate samples')
ax1.set_ylabel('Observed values')
ax1.spines[['right', 'top']].set_visible(False)  
ax1.grid(which='minor', color='#d6d6d6', linestyle=':')
ax1.grid(which='major', color='#949494', linestyle='--')
plt.legend()
#plt.savefig('Frequency Distribution of Minimum Distance to the Boundary_.png', dpi=400, bbox_inches="tight", pad_inches=0.1)

plt.show()
```
